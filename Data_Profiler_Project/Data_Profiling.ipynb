# Part - D Exploratory Data Analysis (EDA)

Exploratory Data Analysis (EDA) is used to understand the dataset, identify patterns, detect anomalies, and prepare data for further modeling.

### Univariate Analysis

```python
# Univariate Analysis 

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Histogram
axes[0].hist(df['runs'], bins=20, color='#2563eb', edgecolor='white')
axes[0].set_title('Distribution')

# Boxplot
sns.boxplot(x=df['over'], ax=axes[1], color='#7c3aed')
axes[1].set_title('Outliers Detection')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='batting_team', palette='Set2')
plt.title('Category Distribution')
plt.xticks(rotation=45)
plt.show()
```

### Bivariate Analysis

```python
sns.scatterplot(data=df, x='striker_strike_rate', y='total_runs')

plt.title('Strike Rate vs Total Runs')
plt.show()
```

```python
sns.boxplot(data=df, x='batting_team', y='total_runs')

plt.title('Batting Team vs Total Runs')
plt.xticks(rotation=45)

plt.show()
```

### Multivariate Analysis

```python
plt.figure(figsize=(8,6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.set(style="white", font_scale=1.2)

pair = sns.pairplot(
    df.select_dtypes(include='number'),
    diag_kind='kde',          
    corner=True,             
    plot_kws={'alpha': 0.6, 's': 50, 'edgecolor': 'white'},
    diag_kws={'fill': True}
)

pair.fig.set_size_inches(12, 10)   # 🔥 Make it BIG
pair.fig.suptitle("Pairwise Relationships Between Numerical Features", y=1.02)

plt.show()
```
